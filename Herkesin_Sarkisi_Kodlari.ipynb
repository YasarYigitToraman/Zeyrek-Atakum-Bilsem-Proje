```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings("ignore")

oncekiDinlenmeSayisi=0
veri = pd.read_csv("fizydinlenensarkilar.csv")

plot_order = veri.groupby('Sarki Ismi')['Dinlenme Sayisi'].sum().sort_values(ascending=False).index.values

sb.catplot(data=veri, x='Sarki Ismi',  y='Dinlenme Sayisi',kind='bar',errorbar=None, legend_out=False, order=plot_order)
plt.xticks(rotation = 90)
plt.xticks(fontsize=9)
plt.show()

print("En çok dinlenen şarkı", plot_order[0])
```
